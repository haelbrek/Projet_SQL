#### Création des plats.

Un plat comprend : 
- Le pain du burger.
- 1 ou 2 viandes (steak, fish ou nugguet).
- 0 ou 1 sauce.
- De 0 à 3 fromages


````python
df_ingredient_plat = pd.DataFrame({
    "ingredient": ["pain", "steak", "nugguet", "fish", "salade", "tomate", "cornichon", "s_mayo", "s_ketchup", "s_barbecue", "fromage", "sans"], 
    "prix": [0.49, 0.99, 1.19, 1.49, 0.29, 0.29, 0.09, 0.19, 0.19, 0.19, 0.39, 0],
    "stock": [rd.randint(1000, 5000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(200, 2000), rd.randint(800, 4000), 0, ]
    })

df_ingredient_plat
````

````python
nb_plat = 50
list_plat = []

for plat in range (nb_plat):
    rd_viande = rd.randint(1,3)
    nb_viande = rd.randint(1,2)
    rd_crudites = rd.randint(4,7)
    if rd_crudites == 7:
        rd_crudites = 11
    rd_sauce = rd.randint(7,10)
    if rd_sauce == 10:
        rd_sauce = 11
    nb_fromage = rd.randint(0,3)

    list_plat.append({
        "name": "Plat n°" + str(plat + 1),
        "ingredients": [df_ingredient_plat.ingredient[0], df_ingredient_plat.ingredient[rd_viande], df_ingredient_plat.ingredient[rd_crudites], df_ingredient_plat.ingredient[rd_sauce], df_ingredient_plat.ingredient[10]],
        "prix": round((df_ingredient_plat.prix[0] + nb_viande * df_ingredient_plat.prix[rd_viande] + df_ingredient_plat.prix[rd_crudites] + df_ingredient_plat.prix[rd_sauce]) + nb_fromage *  df_ingredient_plat.prix[10] + 1.49, 2),
    })

df_plat = pd.DataFrame(list_plat)

df_plat
````

````python
df_ingredient_plat = df_ingredient_plat.set_index("ingredient")

df_ingredient_plat
````

````python
df_ingredient_plat["stock"]["pain"]
````

````python
df_plat.ingredients[2]
````

````python
def stock(liste = list):
    stock = 10000000000
    for element in liste:
        # print (element)
        if element != "sans":
            if df_ingredient_plat["stock"][element] < stock:
                stock = df_ingredient_plat["stock"][element]
    return stock
````

````python
stock(df_plat.ingredients[2])
````

````python
df_plat["stock"] = 0

df_plat.head(5)
````

Un dessert comprend : 
- Un fruit.
- Entre 1 et 3 boules de glace.
- 1 sauce et des croustilles.


````python
df_fruit.ingredient[1]
````

````python
nb_dessert = 50
list_dessert = []

for dessert in range (nb_dessert):
    rd_fruit = rd.randint(0, len(df_fruit)-1)
    rd_glace = rd.randint(0, len(df_glace)-1)
    rd_sauce = rd.randint(0, len(df_sauce)-1)
    rd_boule = rd.randint(1, 3)

    list_dessert.append({
        "name": "Dessert n°" + str(dessert + 1),
        "ingredients": [df_fruit.ingredient[rd_fruit], df_glace.ingredient[rd_glace], df_sauce.ingredient[rd_sauce], df_sauce.ingredient[3]],
        "prix": round((df_fruit.prix[rd_fruit] + rd_boule * df_glace.prix[rd_glace] + df_sauce.prix[rd_sauce] + df_sauce.prix[3]) + 0.49, 2),
    })

df_dessert = pd.DataFrame(list_dessert)

df_dessert.head(10)
````

````python
df_dessert["stock"] = 0

df_dessert.head(10)
````

````python
df_ingredient_dessert = pd.concat([df_fruit, df_glace, df_sauce]).set_index("ingredient")

df_ingredient_dessert
````

````python
df_dessert["ingredients"][0]
````

````python
def stock_dessert(liste = list):
    stock = 10000000000
    for element in liste:
        # print (element)
        if element != "sans":
            if df_ingredient_dessert["stock"][element] < stock:
                stock = df_ingredient_dessert["stock"][element]
    return stock
````

````python
for element in range(len(df_dessert)):
    df_dessert["stock"][element] = stock_dessert(df_dessert["ingredients"][element])

df_dessert
````
